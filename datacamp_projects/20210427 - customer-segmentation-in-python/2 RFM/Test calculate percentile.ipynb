```python
import pandas as pd
df = pd.read_excel('random number.xlsx')
```

this is the case assigning group number where the greater the better

```python
# range endpoint must be q+1
spend_quartiles = pd.qcut(df['number'],q=7,labels=range(1,8))
data1=df.copy()
data1['Q']=spend_quartiles
data1.sort_values('number')
```

this is the case assigning group number where the smaller the better<br />
example : recency in RFM

```python
spend_quartiles_inverse= pd.qcut(df['number'],q=7,labels=range(7,0,-1))
data2=df.copy()
data2['Q']=spend_quartiles_inverse
data2.sort_values('number')
```

this is the case assigning text as group number

```python
r_labels = ['Active','Lapsed','Inactive','Churned']
spend_quartiles_label= pd.qcut(df['number'],q=4,labels=r_labels)
data3=df.copy()
data3['Q']=spend_quartiles_label
data3.sort_values('number')
```
